```python
import pandas as pd
import numpy as np
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
df.head(10)
df.tail(10)
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style",
           "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type",
          "num-of-cylinders","engine-type","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm",
          "city-mpg","highway-mpg","price"]
df.columns = headers
df.head(10)
```

```python
#to drop the values with Na/NaN , axis- 0, removes row, 1 removes column
#inplace=True saves the changes to df
df.dropna(subset=["price"], axis=0, inplace= True )
df.head(10)
```

```python
#replace ? with NaN
df['normalized-losses'] = df['normalized-losses'].replace('?',np.nan).astype(float)
df["normalized-losses"] = df["normalized-losses"].fillna(value=df["normalized-losses"].mean())
df["normalized-losses"]

df["num-of-doors"] = df["num-of-doors"].replace('?',np.nan)
df["num-of-doors"] = df["num-of-doors"].fillna(value="four")
df["num-of-doors"]

df["bore"] = df["bore"].replace('?',np.nan).astype(float)
df["bore"] = df["bore"].fillna(value=df["bore"].mean())
df["bore"].iloc[50:61]

df["stroke"] = df["stroke"].replace('?',np.nan).astype(float)
df["stroke"] = df["stroke"].fillna(value=df["stroke"].mean())
df["stroke"].iloc[50:61]

df["horsepower"] = df["horsepower"].replace('?',np.nan).astype(float)
df["horsepower"] = df["horsepower"].fillna(value=df["horsepower"].mean())
df["horsepower"].iloc[125:135]

df["peak-rpm"] = df["peak-rpm"].replace('?',np.nan).astype(float)
df["peak-rpm"] = df["peak-rpm"].fillna(value=df["peak-rpm"].mean())
df["peak-rpm"].iloc[125:135]

df["price"] = df["price"].replace('?',np.nan).astype(float)
df["price"] = df["price"].fillna(value=df["price"].mean())

df["price"].iloc[0:15]
```

```python
#Data normalization 
#Simple feature scaling
df["symboling"] = df["symboling"]/df["symboling"].max()
df["symboling"]

#min-max method
df["symboling"] = df["symboling"]-df["symboling"].min()/df["symboling"].max()-df["symboling"].min()
df["symboling"]

#zscore
df["symboling"] = df["symboling"]-df["symboling"].mean()/df["symboling"].std()
df["symboling"]
```

```python
bins = np.linspace(min(df["price"]),max(df["price"]),4)
group_names = ["Low","Medium","High"]
df["price-binned"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
df.head(20)
```
